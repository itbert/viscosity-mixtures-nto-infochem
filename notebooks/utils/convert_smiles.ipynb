```python
import numpy as np
import pandas as pd
from classes.descriptors import Smiles2Descriptors
from rdkit import Chem
```

```python
teo_data = pd.read_csv('data/NTO_smiles_encoded.csv')
teo_data
```

```python
df = pd.DataFrame(teo_data)
```

```python
def calculate_descriptors(smiles):
    try:
        descriptor = Smiles2Descriptors(smiles)
        return (
            descriptor.logp, 
            descriptor.tpsa, 
            descriptor.molwt,
            descriptor.vdw_volume,
            descriptor.fraction_non_rotatable_bounds,
            descriptor.num_atoms
        )
    except ValueError as e:
        print(f"Ошибка при обработке SMILES: {e}")
        return None, None, None, None, None, None
    except Exception as e:
        print(f"Неизвестная ошибка: {e}")
        return None, None, None, None, None, None
```

```python
df["logp"], df["tpsa"], df["molwt"], df['vdw_volume'], df['fraction_non_rotatable_bunds'], df['num_atoms']= zip(*df["smiles"].apply(calculate_descriptors))

# Отображение результата
df
```
